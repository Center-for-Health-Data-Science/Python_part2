# Objects

The result of us performing PCA is a PCA object. It is a sort of bundle that contains several pieces of information, such as the Eigenvalues.

The information pieces attached to any object are called its **attributes**. Here are some of the PCA object's attributes (it has more!):

* `PCA_obj.components_` - the Eigenvectors
* `PCA_obj.n_features_in_` - How many original features there are
* `PCA_obj.explained_variance_` - The variance explained by each of the principal components

Attributes are addressed like so:

`object.attrib`

You can list all of an object's public attributes with `vars()`:

```python
vars(PCA_obj)
```

Some attributes can be private and you cannot display them in this way.
It is very important to remember that attributes are *never* followed by paranthesis! Their name may end in an underscore `_` though (and in scikit-learn this has a special meaning we'll get to later).

The operations we can perform on an object are called **methods**. On a PCA object for example we can perform a `fit_transform`.

In general we can call a method like so:

`object.method(parameters)`

Methods are *always* followed by paranthesis since they get executed.

# Functions

You can think of functions like 'mini-programs' that receive input and do something to it. Often this will result in an output (but not always):

**Input -> [Function processes input] (-> Output)**

We have already used many _built-in_ functions such as `print()` and `int()`.

The good thing about functions is that they **abstract tasks**. For example the `len()` function calculates the length of an object. This object could be of different nature and structure, i.e. a list, a dictionary, a dataframe or a string. From a user perspective, we are not overly interested in _how_ one finds the length of these different data types, only that it gets done.

# Classes

All objects in python are instances of their **class**. Which class an object belongs to defines what _attributes_ it has and what _methods_ we can execute on the object.

The `PCA_obj` we used is an instance of the `PCA` class and we create it like so:

````python
from sklearn.decomposition import PCA
#Init a PCA object
PCA_obj = PCA()
````

````python
type(PCA_obj)
````

Because it is a instance of the `PCA` class is has the **attributes** defined in the PCA class, such as `.components_` and `.explained_variance_`, and we can execute the **methods** defined on the PCA class, such as `fit_transform()`.

Actually methods are just functions which are specific to this object class and defined on it.
